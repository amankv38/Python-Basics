```python
from matplotlib import pyplot as plt
```

```python
a=[1,2,3,4,5,6,7,8,9]
b=[11,22,33,44,55,66,77,88,99]
plt.plot(a,b)
plt.title("INFORMATION")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.show
```

```python
from matplotlib import style
```

```python
style.use('ggplot')
x=[1,2,3,4,5,4,3,2,1]
y=[11,19,18,17,12,14,13,15,16]
plt.plot(x,y,'c',label="AAAAA",linewidth=7)
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.title("GRAPHH")
plt.legend()
plt.grid(True,color='g')

plt.show
```
